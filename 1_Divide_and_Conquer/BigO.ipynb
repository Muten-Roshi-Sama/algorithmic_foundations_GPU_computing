### Example : Linear Time O(n)

```python
def addUp(n):
    sum = 0
    for i in range(n):
        sum += i
    return sum


# n= 100.000
# ~ 100.000 steps ...
```

### Example : Constant Time O(n)

```python
def addUp_fast(n):
    sum = n * (n+1) / 2
    return sum

# n = 100.000
# ~ 3 steps : addition, multiplication, division

# O(1) : result(number of steps taken to solve) independent of n !
```
